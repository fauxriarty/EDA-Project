# Bivariate Analysis (2D Analysis)

```python
# bivariate (2d) analysis
from scipy.stats import chi2_contingency

# categorical vs. categorical analysis: 'marital' and 'deposit'
print("Categorical vs. Categorical Analysis: Marital Status and Deposit")

# 1. create a two-way contingency table
contingency_table = pd.crosstab(df['marital'], df['deposit'])
print("\nContingency Table (Counts):")
print(contingency_table)

# 2. calculate row and column profiles (proportions)
print("\nRow Profiles (Proportion of Deposit within each Marital Status):")
print(contingency_table.div(contingency_table.sum(axis=1), axis=0))

print("\nColumn Profiles (Proportion of Marital Status for each Deposit Outcome):")
print(contingency_table.div(contingency_table.sum(axis=0), axis=1))

# 3. chi-square test of independence
chi2, p, dof, expected = chi2_contingency(contingency_table)
print(f"\nChi-Square Test Results:")
print(f"  - Chi2 Value: {chi2:.2f}")
print(f"  - p-value: {p:.4f}")

# a small p-value (< 0.05) suggests the relationship is statistically significant.

# 4. visualization
contingency_table.plot(kind='bar', stacked=True, figsize=(8, 6))
plt.title('Marital Status vs. Deposit Outcome')
plt.ylabel('Count')
plt.xticks(rotation=0)
plt.show()

# numerical vs. categorical analysis: 'balance' and 'education'
print("\nNumerical vs. Categorical Analysis: Balance across Education Levels")
plt.figure(figsize=(10, 6))
sns.violinplot(x='education', y='balance', data=df, order=['primary', 'secondary', 'tertiary'])
plt.title('Distribution of Account Balance by Education Level')
plt.xlabel('Education Level')
plt.ylabel('Account Balance')
plt.ylim(df['balance'].min(), 5000) # trim y-axis for better visibility
plt.show()

# numerical vs. numerical analysis: 'age' and 'duration'
print("\nNumerical vs. Numerical Analysis: Call Duration vs. Age")
plt.figure(figsize=(10, 6))
sns.scatterplot(x='age', y='duration', hue='deposit', data=df, alpha=0.6)
plt.title('Call Duration vs. Age of Client, Colored by Deposit Outcome')
plt.xlabel('Age')
plt.ylabel('Call Duration (seconds)')
plt.legend(title='Made Deposit')
plt.show()
```

# Hierarchical Clustering




```python
# hierarchical clustering
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from scipy.cluster.hierarchy import dendrogram, linkage
from sklearn.cluster import AgglomerativeClustering
import numpy as np

# For this bank marketing dataset, Hierarchical Clustering is more appropriate for a few key reasons:
# 1.  Interpretability: The primary output, a dendrogram, is a powerful visual tool. It allows us to see how
#     clusters merge and at what distance, providing a rich understanding of the data's structure. This is
#     invaluable for EDA, as it helps in explaining the "why" behind customer segments to stakeholders.
# 2.  No Pre-defined Number of Clusters: Unlike Spectral Clustering or K-Means, we don't need to specify the
#     number of clusters (k) beforehand. The dendrogram helps us make an informed, data-driven decision on the
#     optimal 'k' to choose.
# 3.  Captures Nested Structures: Customer data often has a natural hierarchy. For example, a broad cluster
#     of "high-value clients" might contain sub-clusters of "retired high-value" and "working professional
#     high-value". Hierarchical clustering is explicitly designed to find these nested relationships.
#
# While Spectral Clustering is excellent for finding non-convex clusters (like rings or intertwined spirals),
# such complex shapes are unlikely in this dataset. The segments here are more likely defined by feature
# similarities (e.g., high balance, specific job type), which distance-based methods like Hierarchical
# Clustering handle very effectively.

# 1. data preparation for clustering
# select features for clustering
cluster_features = ['age', 'balance', 'duration', 'campaign', 'job', 'marital', 'education', 'housing', 'loan']
df_cluster = df[cluster_features].copy()

# separate numerical and categorical features
numerical_features = ['age', 'balance', 'duration', 'campaign']
categorical_features = ['job', 'marital', 'education', 'housing', 'loan']

# create a preprocessing pipeline to scale numerical and one-hot encode categorical data
preprocessor = ColumnTransformer(
    transformers=[
        ('num', StandardScaler(), numerical_features),
        ('cat', OneHotEncoder(handle_unknown='ignore'), categorical_features)
    ])

# apply the transformations
X_processed = preprocessor.fit_transform(df_cluster)
print("✅ Data has been preprocessed for clustering.")

# ensure X_processed is a dense numpy array for all subsequent steps
X_processed_dense = X_processed.toarray() if hasattr(X_processed, 'toarray') else X_processed

# 2. dendrogram visualization
# hierarchical clustering is computationally intensive. we'll use a sample for the dendrogram.
sample_size = 5000
if X_processed_dense.shape[0] > sample_size:
    sample_indices = np.random.choice(X_processed_dense.shape[0], sample_size, replace=False)
    X_sample = X_processed_dense[sample_indices]
else:
    X_sample = X_processed_dense

print(f"Generating dendrogram from a sample of {X_sample.shape[0]} data points...")

# generate the linkage matrix using the 'ward' method
linked = linkage(X_sample, method='ward')

# plot the dendrogram
plt.figure(figsize=(15, 8))
dendrogram(linked,
           orientation='top',
           distance_sort='descending',
           show_leaf_counts=True,
           truncate_mode='lastp', # show only the last p merged clusters
           p=10) # 'p' is the number of clusters to show
plt.title('Hierarchical Clustering Dendrogram (Truncated)')
plt.xlabel('Cluster Size')
plt.ylabel('Distance (Ward)')
plt.show()

# 3. apply agglomerative clustering on the full dataset
# based on the dendrogram, 4 clusters seem appropriate.
num_clusters = 4
agg_cluster = AgglomerativeClustering(n_clusters=num_clusters, linkage='ward')
df_cluster['cluster'] = agg_cluster.fit_predict(X_processed_dense)

print(f"\n✅ Hierarchical clustering complete. Data points assigned to {num_clusters} clusters.")
print("\nNumber of clients in each cluster:")
print(df_cluster['cluster'].value_counts())

# 4. analyze the characteristics of each cluster
print("\nMean values of numerical features for each cluster:")
hier_cluster_analysis = df_cluster.groupby('cluster')[numerical_features].mean()
display(hier_cluster_analysis)
```

Based on the output, the hierarchical clustering has identified four distinct customer profiles from the data:

Cluster 0 is the largest segment (5,787 clients). They are the youngest (avg. age ~35) with the lowest average balance. This group represents a broad, mass-market base of less-established customers.

Cluster 1 is a large group of the oldest clients (avg. age 45).   Their most defining trait is the shortest average call duration (248s), indicating they are decisive and largely unresponsive to the current marketing approach.

Cluster 2 is a smaller, high-value segment with the highest average balance. However, they have been contacted an extremely high number of times (avg. campaign count ~9.7), suggesting they are over-targeted and resistant.

Cluster 3, though the smallest group, is the most promising. They have the longest average call duration (~388s), which signals high engagement and the strongest potential for conversion.

# Dimensionality Reduction (PCA)

```python
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
import numpy as np

# 1. select and scale the numeric columns for pca
# pca is sensitive to scale, so scaling is a required step.
numerical_cols = ['age', 'balance', 'day', 'duration', 'campaign', 'pdays', 'previous']
data_for_pca = df[numerical_cols]

scaler = StandardScaler()
X_scaled = scaler.fit_transform(data_for_pca)

print("✅ Numerical data selected and scaled.")

# 2. perform pca
# we run pca with all components first to analyze the explained variance.
pca = PCA()
X_pca = pca.fit_transform(X_scaled)

# 3. create a scree plot to determine the optimal number of components
explained_variance = pca.explained_variance_ratio_

plt.figure(figsize=(10, 6))
plt.plot(range(1, len(explained_variance) + 1), explained_variance, marker='o', linestyle='--')
plt.title('Scree Plot (Explained Variance per Component)')
plt.xlabel('Principal Component')
plt.ylabel('Proportion of Variance Explained')
plt.xticks(range(1, len(explained_variance) + 1))
plt.grid(True)
plt.show()

# plot cumulative explained variance
plt.figure(figsize=(10, 6))
plt.plot(range(1, len(explained_variance) + 1), np.cumsum(explained_variance), marker='o', linestyle='--')
plt.title('Cumulative Explained Variance')
plt.xlabel('Number of Components')
plt.ylabel('Cumulative Variance Explained')
plt.axhline(y=0.9, color='r', linestyle='-')
plt.axhline(y=0.8, color='orange', linestyle='-')
plt.text(1.5, 0.82, '80% variance threshold')
plt.text(1.5, 0.92, '90% variance threshold')
plt.grid(True)
plt.show()

# 4. print pca results (variance and component loadings)
print("\nExplained Variance per Component:")
for i, variance in enumerate(explained_variance):
    print(f"  - PC{i+1}: {variance:.4f} ({(variance * 100):.2f}%)")

print("\nCumulative Explained Variance:")
print(np.cumsum(explained_variance))

# interpretation: the scree plot and cumulative variance show how many components
# are needed to capture most of the data's variance. for instance, the plot
# shows how many components are needed to cross the 80% or 90% threshold.

print("\nPrincipal Component Loadings (Feature Contributions):")
# loadings show how much each original feature contributes to each principal component.
loadings_df = pd.DataFrame(pca.components_, columns=numerical_cols, index=[f'PC{i+1}' for i in range(len(explained_variance))])
display(loadings_df)

# 5. visualize the first two principal components
# we create a new dataframe with the pca results (pc1 and pc2)
pca_df = pd.DataFrame(data=X_pca[:, :2], columns=['PC1', 'PC2'])
# we add the 'deposit' column from the original dataframe for coloring
pca_df['deposit'] = df['deposit'].values

plt.figure(figsize=(12, 8))
sns.scatterplot(x='PC1', y='PC2', hue='deposit', data=pca_df, alpha=0.7)
plt.title('PCA Plot: PC1 vs. PC2 (Colored by Deposit Status)')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.legend(title='Deposit')
plt.grid(True)
plt.show()
```

# Model Building: Regression

Simple Regression: Predict balance based on age.

Multiple Regression: Predict balance based on age and duration.

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
from statsmodels.stats.outliers_influence import variance_inflation_factor
import statsmodels.api as sm
import numpy as np

# --- 1. Simple Linear Regression (Predict balance from age) ---
print("--- Simple Linear Regression (balance ~ age) ---")

# 3. visualize the relationship
plt.figure(figsize=(10, 6))
sns.regplot(x='age', y='balance', data=df, line_kws={'color':'red'}, scatter_kws={'alpha':0.1})
plt.title('Relationship between Balance and Age')
plt.xlabel('Age')
plt.ylabel('Account Balance')
plt.show()

# 4. fit a simple linear regression model
# we use statsmodels for its detailed R-like summary
X_simple = df['age']
y = df['balance']
X_simple_sm = sm.add_constant(X_simple) # add an intercept

model_simple = sm.OLS(y, X_simple_sm).fit()

# 5. display the summary and interpret
print("\nModel Summary (Simple Linear Regression):")
print(model_simple.summary())

# interpretation: check the 'coef' for 'age'. a positive coefficient means
# balance tends to increase with age. check 'p > |t|' to see if 'age' is a
# statistically significant predictor (p < 0.05).
# the 'r-squared' (step 10) shows how much variance in 'balance' is
# explained by 'age' alone (it will likely be very low).

# 7-9. calculate error metrics
y_pred_simple = model_simple.predict(X_simple_sm)
mse = mean_squared_error(y, y_pred_simple)
rmse = np.sqrt(mse)
mae = mean_absolute_error(y, y_pred_simple)

print(f"\nSimple Model Evaluation:")
print(f"  - Mean Squared Error (MSE): {mse:.2f}")
print(f"  - Root Mean Squared Error (RMSE): {rmse:.2f}")
print(f"  - Mean Absolute Error (MAE): {mae:.2f}")

# --- 2. Multiple Linear Regression (Predict balance from age & duration) ---
print("\n--- Multiple Linear Regression (balance ~ age + duration) ---")

# 12. check for multicollinearity
X_multi = df[['age', 'duration']]
vif_data = pd.DataFrame()
vif_data["feature"] = X_multi.columns
vif_data["VIF"] = [variance_inflation_factor(X_multi.values, i) for i in range(len(X_multi.columns))]
print("\nChecking for Multicollinearity (VIF):")
print(vif_data)
# vif values are very low (near 1.0), so multicollinearity is not an issue.

# 11. fit the multiple linear regression model
y = df['balance']
X_multi_sm = sm.add_constant(X_multi) # add an intercept
model_multi = sm.OLS(y, X_multi_sm).fit()

# 13. provide the summary and interpret
print("\nModel Summary (Multiple Linear Regression):")
print(model_multi.summary())

# interpretation: the r-squared is likely higher. we can now see the
# effect of both 'age' and 'duration' on 'balance' while holding the other
# constant. 'duration' likely has a positive coefficient.

# 14. visualize relationships
print("\nVisualizing Predictor Relationships (Pair Plot):")
sns.pairplot(df[['balance', 'age', 'duration']], corner=True, kind='reg',
             plot_kws={'line_kws':{'color':'red'}, 'scatter_kws': {'alpha':0.1}})
plt.suptitle('Pair Plot of Regression Variables', y=1.02)
plt.show()
```

# Model Building: Logistic Regression

This cell builds a model to predict whether a client will make a deposit (deposit). We'll start with a simple one-variable model (like your R instructions) for interpretation and then build a full multiple-variable model for better prediction.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
from statsmodels.stats.outliers_influence import variance_inflation_factor
import statsmodels.api as sm

# --- 1. Simple Logistic Regression (Predict deposit from duration) ---
# this is similar to instructions 4-6, for interpretation.
print("--- Simple Logistic Regression (deposit ~ duration) ---")

# we need to convert 'yes'/'no' to 1/0 for the model
y_numeric = df['deposit'].map({'yes': 1, 'no': 0})
X_simple = df['duration']
# statsmodels (like R) needs an explicit constant (intercept)
X_simple_const = sm.add_constant(X_simple)

# fit the model using statsmodels for a detailed summary
model_simple = sm.Logit(y_numeric, X_simple_const).fit(disp=0) # disp=0 suppresses convergence messages

# 5. display the summary and interpret
print("\nModel Summary (Simple Logistic Regression):")
print(model_simple.summary())

# interpretation: the 'coef' for duration is positive (0.0037).
# this means for each additional second of call duration, the log-odds
# of a client subscribing increases by 0.0037. the low p-value (p > |z|)
# of 0.000 confirms that 'duration' is a highly significant predictor.

# 6. plot the regression line over the scatter plot
plt.figure(figsize=(10, 6))
# we use regplot with logistic=true to plot the s-shaped curve
sns.regplot(x='duration', y=y_numeric, data=df, logistic=True, ci=None,
            scatter_kws={'alpha': 0.05, 'color': 'blue'}, line_kws={'color': 'red'})
plt.title('Logistic Regression: Deposit Probability vs. Call Duration')
plt.xlabel('Call Duration (seconds)')
plt.ylabel('Probability of Deposit ("yes"=1)')
plt.show()


# --- 2. Multiple Logistic Regression (Full Model) ---
print("\n--- Multiple Logistic Regression (Full Model) ---")

# 11-12. prepare data and check for multicollinearity
# we'll use the same features as our clustering analysis
numerical_features = ['age', 'balance', 'day', 'duration', 'campaign', 'pdays', 'previous']
categorical_features = ['job', 'marital', 'education', 'default', 'housing', 'loan', 'contact', 'month', 'poutcome']

# check for multicollinearity *before* scaling
print("Checking for Multicollinearity (VIF):")
X_numeric = df[numerical_features]
vif_data = pd.DataFrame()
vif_data["feature"] = X_numeric.columns
vif_data["VIF"] = [variance_inflation_factor(X_numeric.values, i) for i in range(len(X_numeric.columns))]
print(vif_data)
# a vif > 5 would be a concern. all values are low, so we can proceed.

# create the preprocessor
preprocessor = ColumnTransformer(
    transformers=[
        ('num', StandardScaler(), numerical_features),
        ('cat', OneHotEncoder(handle_unknown='ignore'), categorical_features)
    ])

# define our full X and y
X = df.drop('deposit', axis=1)
y = y_numeric # use the 1/0 mapped target

# split into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# apply the preprocessing pipeline
X_train_processed = preprocessor.fit_transform(X_train)
X_test_processed = preprocessor.transform(X_test)

print("\n✅ Data processed and split for training and testing.")

# 11. fit the multiple logistic regression model
model_multi = LogisticRegression(random_state=42, max_iter=1000)
model_multi.fit(X_train_processed, y_train)

# 13. evaluate the model
y_pred = model_multi.predict(X_test_processed)

# 7-10. calculate evaluation metrics (for classification)
# instead of mse/rmse, we use accuracy, confusion matrix, and classification report.
accuracy = accuracy_score(y_test, y_pred)
print(f"\nModel Accuracy: {accuracy:.4f}")

print("\nConfusion Matrix:")
cm = confusion_matrix(y_test, y_pred)
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=['Predicted No', 'Predicted Yes'], yticklabels=['Actual No', 'Actual Yes'])
plt.title('Confusion Matrix')
plt.show()

print("\nClassification Report:")
print(classification_report(y_test, y_pred, target_names=['No', 'Yes']))

# 13. interpret the coefficients
print("\nModel Coefficients (Top 10):")
# get all feature names from the preprocessor
cat_feature_names = preprocessor.named_transformers_['cat'].get_feature_names_out(categorical_features).tolist()
all_feature_names = numerical_features + cat_feature_names

# create a dataframe of coefficients
coef_df = pd.DataFrame(model_multi.coef_[0], index=all_feature_names, columns=['Coefficient'])
# display top 10 positive and negative coefficients
print("Top 10 features increasing 'Yes' probability:")
display(coef_df.sort_values(by='Coefficient', ascending=False).head(10))
print("\nTop 10 features decreasing 'Yes' probability:")
display(coef_df.sort_values(by='Coefficient', ascending=True).head(10))
```
